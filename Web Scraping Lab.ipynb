```python
#Get the items from each restaurant
def get_restaurants_items(name_rest_fixed):
    url = 'http://chubbygrub.com/{}/'.format(name_rest_fixed)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    rest_items = soup.find('table')
    table = rest_items
    items = []
    for row in table.find('tbody').find_all('tr'):
        item={}
        cells=row.find_all('td')
        item['name']=cells[0].text
        item['category'] =cells[1].text.split()
        item['calories']=cells[2].text
        item['fats'] =cells[3].text
        item['carbs']=cells[4].text
        item['url handles']=name_rest_fixed
        
        items.append(item)
    pf=pd.DataFrame(items)  
    return pf
        
        
       
```
